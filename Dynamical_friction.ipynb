```python
import numpy as np
import math
from math import *
from astropy import units as u
from astropy import constants as c
from matplotlib import pylab as plt
from astropy.cosmology import Planck15 as cosmo
```

### Dynamical Friction decay time

```python
# Mass_ratio = M_host/M_sat
mass_ratio = np.arange(1.1,50,0.01)
```

```python
H=(1/(cosmo.H(0).value*1000/(1e6*3.085e16)*(365.25*24*3600)))/1e9*u.Gyr
t_df = 1.17*mass_ratio/np.log(mass_ratio)/10*H
```

```python
plt.xkcd()
plt.plot(mass_ratio, t_df)
plt.hlines(H.value*2/3, 0,50,linestyle='dashed')
plt.vlines(16, 0,20,linestyle='dashed')
plt.hlines(H.value, 0,50,linestyle='dotted')
plt.vlines(28, 0,20,linestyle='dotted')
plt.xlabel("Mh/Ms")
plt.ylabel("Gyr")
plt.xlim(1.5,40)
plt.ylim(0,20)
plt.show()
```
